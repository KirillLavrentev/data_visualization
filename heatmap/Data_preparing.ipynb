```python
"""
Это текстовый файл с несколькими столбцами. Первые два столбца - это X и У в метрах. 
X надо поделить на 90000 и вычесть из частного 121, а Y - поделить на 112000 и к результату прибавить 31. 
Тогда получатся долгота и широта, соответственно. 4-8 столбцы - это разные параметры. 
Для тренировки можно взять любой из них. Последний столбец - идентификатор слоя по глубине 
(в массиве всего 7 слоев). Если на конце 1 - это первый слой, если 2 - второй и тд. 
На первую цифру (1 или 2) в этом двузначном числе внимания можно не обращать.

"""
```

```python
import pandas as pd
import numpy as np
```

```python
file = 'Strain01072019.log.out'
b = []
with open(file,'r') as data:
    a = data.readlines()
for i in a:
    b.append(i[1:-1:].split(','))
```

```python
for i in range(len(b)):
    if (len(b[i])!=9):
        print(len(b[i]))
```

```python
"""
Это текстовый файл с несколькими столбцами. Первые два столбца - это X и У в метрах. 
X надо поделить на 90000 и вычесть из частного 121, а Y - поделить на 112000 и к результату прибавить 31. 
Тогда получатся долгота и широта, соответственно. 4-8 столбцы - это разные параметры. 
Для тренировки можно взять любой из них. Последний столбец - идентификатор слоя по глубине 
(в массиве всего 7 слоев). Если на конце 1 - это первый слой, если 2 - второй и тд. 
На первую цифру (1 или 2) в этом двузначном числе внимания можно не обращать.

"""
```

```python
df = pd.DataFrame(b,columns=['longitude', 'latitude', 'param1', 
                            'param2','param3','param4','param5','param6','layer'])
```

```python
df.tail()
```

```python
for i in list(df.columns):
    df[i] = df[i].astype('float')
```

```python
df.tail()
```

```python
df['longitude'] = df['longitude']/90000 - 121
```

```python
df['latitude'] = df['latitude']/112000 + 31
```

```python
df.to_csv('Strain01072019.csv',index=False) # сохраняем обработанные данные 
```
